```python
distance = float(input("거리를 입력하세요 (km): "))
hour = int(input("현재 시간을 입력하세요 (0-23): "))
if distance <= 2:
    result = 3000
else:
    result = 3000 + (distance - 2) * 1000

if hour <= 4 or hour >= 22:
    result = result * 1.2
print(f"요금 : {result}원")
    
```
