```python
number = int(input("숫자를 입력하세요: "))
if number < 2:
    print("2 이상의 숫자만 입력 가능합니다.")
else:
    prime = True
    for i in range(2, number+1):
        if number % i == 0:
            prime = False
            break
    
    if prime:
        print(f"{number}은(는) 소수입니다.")
    else:
        print(f"{number}은(는) 소수가 아닙니다.")

    prime_list = []
    for num in range(2, number+1):
        check = True
        for i in range(2, num):
            if num % i == 0:
                check = False
                break
        if check:
            prime_list.append(num)

    print(f"1부터 {number}까지의 소수: ")
    print(prime_list)
    print(f"총 {len(prime_list)}개의 소수가 있습니다.")
```

```python
year = int(input("년도: "))
month = int(input("월: "))

if month < 1 or month > 12:
    print("잘못된 월입니다.")
else:
    # 1. 1년 1월 1일부터 (year-1)년 12월 31일까지의 총 일수 계산
    total_days = (year - 1) * 365 + (year - 1) // 4 - (year - 1) // 100 + (year - 1) // 400

    # 2. 해당 년도 1월부터 (month-1)월까지의 일수 더하기
    days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        days_in_month[2] = 29

    for i in range(1, month):
        total_days += days_in_month[i]

    # 3. 해당 월의 1일 요일 구하기 (0: 일, 1: 월, ..., 6: 토)
    # 1년 1월 1일이 월요일이므로 +1을 해줍니다.
    start_day = (total_days + 1) % 7

    # 4. 출력
    print(f"\n      {year}년 {month}월")
    print(" 일 월 화 수 목 금 토")

    # 1일이 시작되기 전까지 공백 출력
    for _ in range(start_day):
        print("   ", end="")

    # 날짜 출력
    for day in range(1, days_in_month[month] + 1):
        print(f"{day:2} ", end="")
        
        # 요일 칸이 꽉 차면(토요일이 지나면) 줄바꿈
        if (start_day + day) % 7 == 0:
            print()
    print("\n")
```

```python
students = int(input("학생 수를 입력하세요: "))

if students < 3:
    print("최소 3명 이상의 학생 정보를 입력해야합니다.")

names = []
Koreans = []
Englishs = []
maths = []
avgs = []

for i in range(1, students + 1):
    print(f"[{i}번 학생]")
    name = input("이름: ")
    Korean = int(input("국어: "))
    English = int(input("영어: "))
    math = int(input("수학: "))
    avg = (Korean + English + math) / 3
    
    print(f"국어: {Korean}")
    print(f"영어: {English}")
    print(f"수학: {math}")

    if avg >= 90:
        grade = "A"
    elif avg >= 80:
        grade = "B"
    elif avg >= 70:
        grade = "C"
    elif avg >= 60:
        grade = "D"
    else:
        grade = "F"

    print(f"평균: {avg:.2f}점 (등급 {grade})")

    names.append(name)
    Koreans.append(Korean)
    Englishs.append(English)
    maths.append(math)
    avgs.append(avg)


max_avg = max(avgs)
min_avg = min(avgs)
max_student = names[avgs.index(max_avg)]
min_student = names[avgs.index(min_avg)]

kor_avg = sum(Koreans) / students
eng_avg = sum(Englishs) / students
math_avg = sum(maths) / students

print("=== 전체 통계 ===")
print(f"최고 점수: {max_student} ({max_avg:.2f}점)")
print(f"최저 점수: {min_student} ({min_avg:.2f}점)")
print(f"국어 평균: {kor_avg:.2f}점")
print(f"영어 평균: {eng_avg:.2f}점")
print(f"수학 평균: {math_avg:.2f}점")
```

```python
products = ["니트", "청바지", "후드집업", "모자", "안경", "티셔츠"]
prices = [35000, 25000, 50000, 10000, 5000, 15000]

cart =[]
cart_prices = []

print("=== 쇼핑몰 ===")

while True:
    print("1. 상품목록 2.장바구니담기 3.장바구니보기 4.결제 5.종료")
    choice = int(input("원하시는 메뉴를 선택하세요: "))

    if choice == 1:
        print("--- 상품목록 ---")
        for i in range(len(products)):
            print(f"{i+1}. {products}: {prices}원")
    
    elif choice == 2:
        pick = int(input("담을 상품 번호를 입력하세요: "))
        if 1 <= pick <= len(products):
            cart.append(products[pick-1])
            cart_prices.append(prices[pick-1])
            print(f"{products[pick-1]}를 장바구니에 담았습니다.")
        else:
            print("잘못된 상품 번호입니다.")

    elif choice == 3:
        if not cart:
            print("장바구니가 비어 있습니다.")
        else:
            print(f"장바구니: {len(cart)}개")
            for i in cart:
                print(f"{i}")
            print(f"현재 합계: {sum(cart_prices)}원")

    elif choice == 4:
        total = sum(cart_prices)
        if total == 0:
            print("결제할 상품이 없습니다.")
            continue
        
        print(f"주문 금액: {total}원")

        coupon = input("할인 쿠폰이 있습니까? (y / n): ")
        if coupon.lower() == "y":
            total = total * 0.9
            print("10% 할인이 적용되었습니다.")

        shipping = 0
        if total < 30000:
            shipping = 3000
            print("구매 금액이 30,000원 미만으로 배송비 3,000원이 추가됩니다.")
        else:
            print("30,000원 이상 구매로 배송비가 무료입니다!")

        print(f"최종 결제 금액: {int(total + shipping)}원")
        print("결제가 완료되었습니다. 감사합니다!")
        break

    elif choice == 5:
        print("쇼핑을 종료합니다.")
        break
    else:
        print("잘못된 입력입니다.")

```

```python
def calculate(a, b, operator):
    if operator == "+":
        return a + b
    elif operator == "-":
        return a - b
    elif operator == "*":
        return a * b
    elif operator == "/":
        if b == 0:
            return "0으로 나눌 수 없습니다."
        return a / b
    else:
        return "잘못된 연산자입니다."

print(calculate(10, 3, "+"))
print(calculate(10, 3, "-"))
print(calculate(10, 3, "*"))
print(calculate(10, 3, "/"))
print(calculate(10, 0, "/"))
print(calculate(10, 3, "%"))
```

```python
def get_average(*args):
    if not args:
        return 0
    total = sum(args)
    average = total / len(args)
    return round(average, 2)

print(get_average(80, 90, 100))
print(get_average(75, 82, 91, 68, 95))
print(get_average(100))
print(get_average())
        
```

```python
def print_gugudan(dan, start=1, end=9):
    if start > end:
        print("잘못된 범위입니다")
    for i in range(start, end+1):
        print(f"{dan} x {i} = {start * i}")
        i += 1
    print()


print_gugudan(5)
print_gugudan(3, 3, 7)
print_gugudan(7, 9, 2)
```

```python
def factorial(n):
    total = 1
    if n == 0:
        return 1
    for i in range(1, n+1):
        total = total * i
    return total

print(factorial(5))
print(factorial(0))
```

```python
def list_sum(numbers):
    total = 0
    for number in numbers:
        total += number

    return total

print(list_sum([1, 2, 3, 4 ,5]))
print(list_sum([10, 20, 30]))
```

```python
def my_len(text):
    ls = list(text)
    num = 0
    for i in ls:
        num += 1
    return num

print(my_len("python"))
print(my_len("안녕하세요"))
```

```python
def count_char(text, ch):
    count = 0
    for char in text:
        if char == ch:
            count += 1
    return count
print(count_char("banana", "a"))
print(count_char("hello", "l"))
```

```python
def reverse_text(text):
    reversed_text = ""
    for char in text:
        reversed_text = char + reversed_text
    
    return reversed_text
print(reverse_text("python"))
print(reverse_text("abc"))
```
