```python
class Product:
    """상품 정보를 관리하는 클래스"""
    
    def __init__(self, product_id, name, price, weight, stock):
        """
        Parameters:
        product_id (str): 상품 ID
        name (str): 상품명
        price (int): 가격
        weight (float): 무게(kg)
        stock (int): 재고 수량
        """
        self.product_id = product_id
        self.name = name
        self.price = price
        self.weight = weight
        self.stock = stock
        pass

    def get_info(self):
        return {
            "ID" : self.product_id, 
            "Name" : self.name, 
            "Price" : self.price, 
            "Weight" : self.weight, 
            "Stock" : self.stock
                }

    def update_stock(self, quantity):
        if self.stock < quantity:
            return False
        self.stock -= quantity
        return True        
        
    def apply_discount(self, discount_rate):
        result = self.price * (1 - discount_rate / 100)
        return result
    

class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_product(self, product, quantity):
        if product.update_stock(quantity):
            self.items.append({'product' : product, 'quantity' : quantity})
            return True
        else:
            print(f"{product.name} 재고가 부족하여 추가할 수 없습니다.")
            return False
        
    def remove_product(self, product_id):
        for item in self.items:
            if item['product'].product_id == product_id:
                self.items.remove(item)
                return True
        return False
        
    def get_total_price(self):
        total = 0
        for item in self.items:
            total += item['product'].price * item['quantity']
        return total

    def get_total_weight(self):
        total_w = 0
        for item in self.items:
            total_w += item['product'].weight * item['quantity']
        return total_w

    def show_cart(self):
        print("======== [ 장바구니 목록 ] ========")
        for item in self.items:
            p = item['product']
            print(f"{p.name}: {item['quantity']}개 (개당 {p.price}원)")
        print(f"총 금액: {self.get_total_price()}원")

class Order:
    def __init__(self, order_id, cart, customer_name):
        self.order_id = order_id
        self.cart = cart
        self.customer_name = customer_name
        self.shipping_fee = 3000

    def calculate_final_price(self):
        product_price = self.cart.get_total_price()
        if product_price > 100000:
            self.shipping_fee = 0
        return product_price + self.shipping_fee

    def print_receipt(self):
        print(f"주문 번호: {self.order_id}")
        print(f"고 객 명: {self.customer_name}")
        print("======================================")
        print(" [주문 상품 목록]")
        for item in self.cart.items:
            p = item['product']
            print(f" - {p.name:10}: {item['quantity']}개 | {p.price * item['quantity']:,}원")
        print(f" 상품 합계: {self.cart.get_total_price():,}원")
        print(f" 배송비: {self.shipping_fee:,}원")
        print(f" 최종 결제 금액: {self.calculate_final_price():,}원")

# 상품 생성
laptop = Product("P001", "노트북", 1200000, 2.5, 10)
mouse = Product("P002", "무선마우스", 35000, 0.2, 50)
keyboard = Product("P003", "기계식키보드", 89000, 1.0, 30)

# 장바구니 생성 및 상품 추가
cart = ShoppingCart()
cart.add_product(laptop, 1)
cart.add_product(mouse, 2)
cart.add_product(keyboard, 1)

# 장바구니 확인
cart.show_cart()

# 주문 생성
order = Order("ORD20260211001", cart, "홍길동")

# 주문서 출력
order.print_receipt()
```
