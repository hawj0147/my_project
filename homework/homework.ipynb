```python
school = {
    '1반': {
        '김철수': {'국어': 85, '영어': 90, '수학': 78},
        '이영희': {'국어': 92, '영어': 88, '수학': 95}
    },
    '2반': {
        '박민수': {'국어': 88, '영어': 85, '수학': 90},
        '최지영': {'국어': 95, '영어': 92, '수학': 88}
    }
}
print(f"'1반' 김철수의 수학 점수: {school['1반']['김철수']['수학']}")
score = school['2반']['박민수']['국어'] + school['2반']['박민수']['영어'] + school['2반']['박민수']['수학']
avg = round(score / 3 , 2)
print("'2반' 박민수의 평균 점수:", avg)
```

```python
a_class = {'김철수', '이영희', '박민수', '최지영', '홍길동'}
b_class = {'박민수', '최지영', '정수민', '강호동', '이영희'}

print(a_class.intersection(b_class))
print(a_class & b_class)
print(a_class.union(b_class))
print(a_class | b_class)
print(a_class.difference(b_class))
print(a_class - b_class)
print(b_class.difference(a_class))
print(b_class - a_class)
```
